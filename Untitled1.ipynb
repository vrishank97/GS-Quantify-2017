```python
import pandas as pd
%matplotlib inline
import matplotlib as plt
import numpy as np

train = pd.read_csv("gcTrianingSet.csv")
test= pd.read_csv("gcPredictionFile.csv")
```

```python
train = train[['initialUsedMemory', 'initialFreeMemory', 'query token','cpuTimeTaken','gcRun']]
```

```python
X_train=train.drop(['gcRun'], axis=1)
```

```python
y_train=train['gcRun']
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
X_train=pd.get_dummies(X_train)
clf=GradientBoostingClassifier(n_estimators=10)
scores = cross_val_score(clf, X_train, y_train, cv=3, verbose=50)
```

```python
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_train, y_train, test_size=0.33)
```

```python
clf.fit(X_train1, y_train1)
y_preds1=clf.predict(X_test1)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test1, y_preds1)
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import RidgeClassifier
```

```python
from sklearn.model_selection import GridSearchCV

clf=RidgeClassifier()
clf.fit(X_train1, y_train1)
y_preds1=clf.predict(X_test1)
print confusion_matrix(y_test1, y_preds1)
```
